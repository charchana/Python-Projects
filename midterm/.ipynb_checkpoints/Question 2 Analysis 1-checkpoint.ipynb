## Using Article Search API, find during which year there was high rate of voilent crime happenings in the USA

```python
import glob
import os
from operator import itemgetter
from os.path import isfile,isdir
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as pyplot
from pylab import *
import csv
import json
```

```python
path = os.getcwd()
dataFolder = os.path.join(path,'Question 2_Data\Crime')
```

```python
crimeFiles = glob.glob(os.path.join(dataFolder, '*.json'))
```

```python
def dataToWriteInCsv(crimeFiles):
    textsToSearch = {}
    wordSearch = ['terrorist', 'terrorism', 'gun', 'bomb', 'attacks', 'voilence', 'extremist', 'rape', 'racism','armed']
    tokenizer = RegexpTokenizer(r'\w+')
    texts = []
    for file in crimeFiles:
         with open(file) as json_file:
            json_data = json.load(json_file)
            response = json_data['response']
            docs = response['docs']
            for article in docs:
                pubDate = article["pub_date"]
                year = pubDate[0:4]
                if article['lead_paragraph'] != None:
                    lead_paragraph = tokenizer.tokenize(article["lead_paragraph"].lower())
                    texts.append(lead_paragraph)
                if article['abstract'] != None:
                    abstract = tokenizer.tokenize(article["abstract"].lower())
                    texts.append(abstract)
                for keys in article["keywords"]:
                    value = keys['value'].lower()
                    texts.append(value)
                for words in wordSearch:
                    if words in texts:
                        if year not in textsToSearch:
                            textsToSearch[year] = 1
                        else:
                            textsToSearch[year] += 1
    return textsToSearch
```

```python
# function writing the csv file with restuarants details
def csvCreation(filename, crimeData):
    with open(filename, 'a',encoding='utf_8', errors='ignore') as outcsv: # opening a csv file as english file
        #configure writer to write standard csv file
        writer = csv.writer(outcsv, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL, lineterminator='\n')
        # writing the headers for the file
        writer.writerow(['YEAR', 'CRIMES HAPPENED'])
        for eachRes in crimeData.items():
            # writing the data row by row
            writer.writerow([eachRes[0],eachRes[1]])
```

```python
# check whether the directory is already created if not this function creates it
def dir_check(file_path):
    if not os.path.exists(file_path):
        os.makedirs(file_path) # creates the complete folder structure
```

```python
# creating the csv file at specified path
def output(csvPath):
    if os.path.exists(csvPath):
        os.remove(csvPath) # removing the file if already present
        crimeData = dataToWriteInCsv(crimeFiles) # creating the restuarants data
        csvCreation(csvPath, crimeData) # creating the csv file
    else:
        crimeData = dataToWriteInCsv(crimeFiles)  # creating the restuarants data
        csvCreation(csvPath, crimeData) # creating the csv file
```

```python
csvFolder = os.path.join(path,'OutputFiles')
dir_check(csvFolder)
csvPath = os.path.join(csvFolder,'Q2Analysis1_VoilentCrimes.csv') # path is taken
```

```python
output(csvPath) # calling the function to create csv file
textsToSearch = dataToWriteInCsv(crimeFiles)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
def pieChart(textsToSearch):
    count = []
    for keys in textsToSearch.items():
        values = keys[1]
        count.append(values)

    # make a square figure and axes
    figure(1, figsize=(6,6))
    ax = axes([0.1, 0.1, 0.8, 0.8])

    # The slices will be ordered and plotted counter-clockwise.
    labels =  ['2014', '2016', '2015']
    explode=[0, 0, 0.08]

    pie(count, explode=explode, labels=labels,autopct='%1.1f%%', shadow=True, startangle=90)
                    # The default startangle is 0, which would start
                    # the Frogs slice on the x-axis.  With startangle=90,
                    # everything is rotated counter-clockwise by 90 degrees,
                    # so the plotting starts on the positive y-axis.

    title('Voilent crime occurences in the USA for 3 years [2014-2016]', bbox={'facecolor':'0.8', 'pad':5})

    show()
```

```python
pieChart(textsToSearch)
```
