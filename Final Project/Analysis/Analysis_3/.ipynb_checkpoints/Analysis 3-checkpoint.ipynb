# Genres Released for 2012-2015 years

```python
import os
import pandas as pd
import calendar
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation
import seaborn as sns
import plotly.plotly as py
from plotly.graph_objs import *
import matplotlib.pyplot as pyplot
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
from matplotlib.colors import rgb2hex
from matplotlib.patches import Polygon
```

```python
path = os.getcwd() # getting the path of current working directory to create the folder structure
configfile = os.path.join(path + '\Data\IMDB_Movies.csv') # getting the path of the json files to be read
```

```python
all_movies = pd.DataFrame.from_csv(configfile, encoding = "ISO-8859-1").reset_index()
#all_movies = pd.DataFrame(all_movies['Country']=='USA')
all_movies = pd.DataFrame(all_movies, columns=['IMDB_ID', 'Year', 'Genres'])
all_movies = all_movies[all_movies['Genres'] != "Null"]
all_movies.head()
```

```python
lst_col = 'Genres'
x = all_movies.assign(**{lst_col:all_movies[lst_col].str.split('|')})
x.head()
```

```python
genresSplit = pd.DataFrame({col:np.repeat(x[col].values, x[lst_col].str.len())
          for col in x.columns.difference([lst_col])
      }).assign(**{lst_col:np.concatenate(x[lst_col].values)})[x.columns.tolist()]
```

```python
genresList = pd.DataFrame(genresSplit[genresSplit['Year'].notnull()])
```

```python
genresList['Year'] =genresList['Year'].astype(int)
```

```python
genresList['Genres'].unique()
```

```python
listOfGenres = ['Crime', 'Mystery', 'War', 'Action', 'Thriller', 'Adult']
filterMoviesWithGenre =genresList[genresList['Genres'].isin(listOfGenres)]
filterMoviesWithGenre.head()
```

```python
filterMoviesWithGenre['Count'] = 1
filterMoviesWithGenre.head()
```

```python
yearWiseGenres = filterMoviesWithGenre.groupby(['Year', 'Genres'])['Count'].sum().reset_index()
```

```python
yearWiseGenresCount = yearWiseGenres[yearWiseGenres['Count']>10]
yearWiseGenresYears = yearWiseGenresCount[(yearWiseGenresCount['Year']>=2012) & (yearWiseGenresCount['Year']<=2017)].reset_index()
yearWiseGenresYear2012 = yearWiseGenresYears[yearWiseGenresYears['Year']==2012]
yearWiseGenresYear2013 = yearWiseGenresYears[yearWiseGenresYears['Year']==2013]
yearWiseGenresYear2014 = yearWiseGenresYears[yearWiseGenresYears['Year']==2014]
yearWiseGenresYear2015 = yearWiseGenresYears[yearWiseGenresYears['Year']==2015]
```

```python
labels2012 = yearWiseGenresYear2012['Genres'] # x axis for first graph
labels2013 = yearWiseGenresYear2013['Genres'] # x axis for first graph
labels2014 = yearWiseGenresYear2014['Genres'] # x axis for first graph
labels2015 = yearWiseGenresYear2015['Genres'] # x axis for first graph
year2012 = yearWiseGenresYear2012['Count']
year2013 = yearWiseGenresYear2013['Count']
year2014 = yearWiseGenresYear2014['Count']
year2015 = yearWiseGenresYear2015['Count']
# make a square figure and axes
# create a figure with two subplots

explode2012=[0.0, 0.0, 0.08]
explode2013=[0.0,0.0,0.0,0.08]
explode2014 =[0.0,0.0,0.0,0.08,0.0]

fig = plt.figure()

# First Plot
fig.add_subplot(2,2,1)
plt.axis('equal') 
plt.pie(year2012, explode=explode2012, labels=labels2012,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("2012")

# Second plot
fig.add_subplot(2,2,2)
plt.axis('equal') 
plt.pie(year2013,explode=explode2013,labels=labels2013,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("2013")

# Thrid plot
fig.add_subplot(2,2,3)
plt.axis('equal') 
plt.pie(year2014, explode=explode2014, labels=labels2014,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("2014")

# Fourth plot
fig.add_subplot(2,2,4)
plt.axis('equal') 
plt.pie(year2015, explode=explode2013, labels=labels2015,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("2015")

plt.show()
```

# Population of the USA 

```python
crimeFiles = os.path.join(path + '\Data\Crime_US_1975_2015.csv') # getting the path of the json files to be read
```

```python
crimes = pd.DataFrame.from_csv(crimeFiles, encoding = "ISO-8859-1").reset_index()
crimes = pd.DataFrame(crimes, columns=['year','State','total_pop', 'homs_sum', 'rape_sum', 'rob_sum','violent_crime'])
crimes = crimes[crimes.notnull()]
```

```python
# aggregating the population columns of each year per city of USA
crimeYearAndPop =  pd.DataFrame(crimes, columns=['year','State', 'total_pop'])
crimeYearAndPop = crimeYearAndPop.groupby(['year','State'])['total_pop'].sum().reset_index()
crimeYearAndPop = crimeYearAndPop.fillna(0.0)
```

```python
crimeYearAndPop = crimeYearAndPop[crimeYearAndPop['year']>=2012]
crimes20122015 = crimeYearAndPop.groupby('year')['total_pop'].sum().reset_index()
crimes20122015
```

```python
# filtering the 4 years of population 2012-2015
crimeYearAndPop2015 = crimeYearAndPop[crimeYearAndPop['year']==2015]
sortedPopulation = pd.DataFrame(crimeYearAndPop2015.sort('total_pop',ascending=False))
sortedPopulation.head()
```

```python
states = ['New Jersey','Rhode Island','Massachusetts','Connecticut','Maryland','New York','Delaware','Florida','Ohio','Pennsylvania', 'Illinois',
'California','Hawaii','Virginia','Michigan','Indiana','North Carolina','Georgia','Tennessee',
'New Hampshire','South Carolina','Louisiana','Kentucky','Wisconsin','Washington','Alabama','Missouri','Texas',
'West Virginia','Vermont','Minnesota','Mississippi','Iowa','Arkansas','Oklahoma','Arizona','Colorado','Maine','Oregon',
'Kansas','Utah','Nebraska','Nevada','Idaho','New Mexico','South Dakota','North Dakota','Montana','Wyoming','Alaska'
]
statesPop2015 = crimeYearAndPop2015['State'].values
statesPop2015Total = crimeYearAndPop2015['total_pop'].values
popDict = {}
for each in states:
    if each not in statesPop2015:
        popDict[each] = 0.0
    else:
        for i,j in zip(statesPop2015,statesPop2015Total):
             popDict[i] = j
```

```python
# Lambert Conformal map of lower 48 states.
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
        projection='lcc',lat_1=33,lat_2=45,lon_0=-95)
# draw state boundaries.
# data from U.S Census Bureau
# http://www.census.gov/geo/www/cob/st2000.html
shp_info = m.readshapefile('st99_d00','states',drawbounds=True)
# population density by state from
# http://en.wikipedia.org/wiki/List_of_U.S._states_by_population_density
# choose a color for each state based on population density.
colors={}
statenames=[]
cmap = plt.cm.hot # use 'hot' colormap
vmin = 0; vmax = 11211585.0 # set range.
for shapedict in m.states_info:
    #print(shapedict)
    statename = shapedict['NAME']
    #print(statename)
    # skip DC and Puerto Rico.
    if statename not in ['District of Columbia','Puerto Rico']:
        pop = popDict[statename]
        # calling colormap with value between 0 and 1 returns
        # rgba value.  Invert color range (hot colors are high
        # population), take sqrt root to spread out colors more.
        colors[statename] = cmap(1.-np.sqrt((pop-vmin)/(vmax-vmin)))[:3]
    statenames.append(statename)
# cycle through state names, color each one.
ax = plt.gca() # get current axes instance
for nshape,seg in enumerate(m.states):
    # skip DC and Puerto Rico.
    if statenames[nshape] not in ['District of Columbia','Puerto Rico']:
        #print('entering')
        color = rgb2hex(colors[statenames[nshape]]) 
        poly = Polygon(seg,facecolor=color,edgecolor=color)
        ax.add_patch(poly)
plt.title('USA States Population density for the year 2015')
plt.show()
```

```python
year = crimes20122015['year']
population = crimes20122015['total_pop']
xlabels = ['2012','2013','2014','2015']
fig, ax = plt.subplots(1,1) 
ax.plot(year,population,marker='o', linestyle='--', color='g')
# Set number of ticks for x-axis
ax.set_xticks(year)
# Set number of ticks for y-axis
ax.set_yticks(population)
# Set ticks labels for x-axis
ax.set_xticklabels(xlabels, rotation=45, fontsize=12)
title('Increase in population in USA from 2012-2015', fontsize=14)
xlabel('Years')
ylabel('Population (Million)')
show()
```

# Rate of crimes happening in years 2012-2015

```python
crimes = crimes[crimes['year']>=2012]
crimes = crimes.fillna(0.0)
crimes.head()
```

```python
columns = ['year','State', 'homs_sum', 'rape_sum', 'rob_sum', 'violent_crime'] 
cities = pd.DataFrame(crimes, columns=columns)
cities['year'] = cities['year'].astype(str)
cities['totalCrimes'] = cities['homs_sum'] + cities['rape_sum'] + cities['rob_sum'] + cities['violent_crime']
#cities['totalCrimes'] = pd.DataFrame(cities.groupby('department_name')['homs_sum', 'rape_sum', 'rob_sum', 'violent_crime'].sum(axis=1).rest_index())
citiesTotalCrimes = pd.DataFrame(cities[['State', 'totalCrimes']])
```

```python
citiesGroup = cities.groupby(['year','State'])['homs_sum', 'rape_sum', 'rob_sum', 'violent_crime'].sum().reset_index()
citiesGroup2015 = citiesGroup[citiesGroup['year']=='2015']
citiesGroup2015['totalCrimes'] = citiesGroup2015['homs_sum'] + citiesGroup2015['rape_sum'] + citiesGroup2015['rob_sum'] + citiesGroup2015['violent_crime']
cities2015TotalCrimes = pd.DataFrame(citiesGroup2015[['State', 'totalCrimes']])
cities2015TotalCrimes.head()

#cities2015TotalCrimes.sort('totalCrimes',ascending=False)
```

```python
states2015 = cities2015TotalCrimes['State'].values
citiesTotalCrimesTotal = cities2015TotalCrimes['totalCrimes'].values
citiesDict = {}
for each in states:
    if each not in states2015:
        citiesDict[each] = 0.0
    else:
        for i,j in zip(states2015,citiesTotalCrimesTotal):
             citiesDict[i] = j
```

```python
# Lambert Conformal map of lower 48 states.
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
        projection='lcc',lat_1=33,lat_2=45,lon_0=-95)
# draw state boundaries.
# data from U.S Census Bureau
# http://www.census.gov/geo/www/cob/st2000.html
shp_info = m.readshapefile('st99_d00','states',drawbounds=True)
# population density by state from
# http://en.wikipedia.org/wiki/List_of_U.S._states_by_population_density
# choose a color for each state based on population density.
colors={}
statenames=[]
cmap = plt.cm.hot # use 'hot' colormap
vmin = 0; vmax = 89377 # set range.
for shapedict in m.states_info:
    #print(shapedict)
    statename = shapedict['NAME']
    #print(statename)
    # skip DC and Puerto Rico.
    if statename not in ['District of Columbia','Puerto Rico']:
        pop = citiesDict[statename]
        # calling colormap with value between 0 and 1 returns
        # rgba value.  Invert color range (hot colors are high
        # population), take sqrt root to spread out colors more.
        colors[statename] = cmap(1.-np.sqrt((pop-vmin)/(vmax-vmin)))[:3]
    statenames.append(statename)
# cycle through state names, color each one.
ax = plt.gca() # get current axes instance
for nshape,seg in enumerate(m.states):
    # skip DC and Puerto Rico.
    if statenames[nshape] not in ['District of Columbia','Puerto Rico']:
        #print('entering')
        color = rgb2hex(colors[statenames[nshape]]) 
        poly = Polygon(seg,facecolor=color,edgecolor=color)
        ax.add_patch(poly)
plt.title('USA States filled by Crime Rate Density for year 2015')
plt.show()
```

```python
crimesRate=pd.DataFrame(crimes.groupby(['year'])['homs_sum','rape_sum','rob_sum','violent_crime'].sum().reset_index())
crimesRate
```

```python
# plotting the rate of increase or decrease in crimes
homs_sum = crimesRate['homs_sum']
rape_sum = crimesRate['rape_sum']
rob_sum = crimesRate['rob_sum']
violent_crime = crimesRate['violent_crime']
years = ['2012','2013','2014','2015']
x = np.arange(len(years))
bar_width = 0.2
opacity = 0.5 #not so dark
plt.bar(x, homs_sum,bar_width, color='cyan', label='Homicide', alpha=opacity)
plt.bar(x+bar_width, rape_sum,bar_width, color='blue', label='Rape', alpha=opacity)
plt.bar(x+bar_width+bar_width, rob_sum,bar_width, color='green', label='Robbery', alpha=opacity)
plt.bar(x+bar_width+bar_width+bar_width, violent_crime,bar_width, color='red', label='Voilent', alpha=opacity)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.05),
          ncol=3, fancybox=True, shadow=True)
plt.xlabel('Year')
plt.ylabel('Crimes Rate')
plt.title("Number of Crimes for different categories in USA [2012-2015]")
plt.xticks(x+bar_width,years)
plt.tight_layout()
plt.show()
```

# Influence of Movies on the increase of voilent crimes in the USA

```python
# Genres for each year 
yearWiseGenresYears.head()
```

```python
# selecting the crime genre alone
crimeGenre = yearWiseGenresYears[yearWiseGenresYears['Genres']=='Crime']
sumsCrimeGenre = crimeGenre['Count'].sum()
averageGenrePercent = crimeGenre['Count']/sumsCrimeGenre
genreCrimePercent = pd.DataFrame(averageGenrePercent*100)
genreCrimePercent['Year'] = crimeGenre['Year']
# selecting the violent crime from FBI crime
voilentCrimesRate = pd.DataFrame(crimesRate, columns=['year', 'violent_crime'])
sums = voilentCrimesRate['violent_crime'].sum()
averagePercent =voilentCrimesRate['violent_crime']/sums
voilentCrimesPercent = pd.DataFrame(averagePercent*100)
voilentCrimesPercent['Year'] = voilentCrimesRate['year']
#voilentCrimesPercent
```

```python
year = genreCrimePercent['Year']
genreCrime = genreCrimePercent['Count']
violentCrimes = voilentCrimesPercent['violent_crime']
xlabels = ['2012','2013','2014','2015']
fig, ax = plt.subplots() 
ax.plot(year,genreCrime,marker='o', linestyle='--', color='g', label="Crime Movies")
ax.plot(year,violentCrimes,marker='o', linestyle='--', color='r', label = "Voilent Crimes")
# Set number of ticks for x-axis
ax.set_xticks(year)
# St number of ticks for y-axis
#ax.set_yticks(violentCrimes)
# Set ticks labels for x-axis
ax.set_xticklabels(xlabels, rotation=45, fontsize=12)
legend = ax.legend(loc='upper left', shadow=True)
xlabel('Years')
ylabel('Crimes')
title('Influence of Movies in Crime Rate')
plt.show()
```
