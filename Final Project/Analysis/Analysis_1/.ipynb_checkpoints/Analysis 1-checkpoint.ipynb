# Number of movies releasing every year [2012-2015]

```python
import os
import pandas as pd
import calendar
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import hist
import glob
import datetime
import urllib.request
```

```python
path = os.getcwd() # getting the path of current working directory to create the folder structure
configfile = os.path.join(path + '\Data\IMDB_Movies.csv') # getting the path of the json files to be read
```

```python
moviesFiles = pd.DataFrame.from_csv(configfile, encoding = "ISO-8859-1").reset_index()
movies = pd.DataFrame(moviesFiles, columns=['Release_Date', 'Rating', 'UrlPoster', 'Title'])
movies = movies[movies['Release_Date'].notnull()]
movies = movies[movies['UrlPoster'].notnull()]
movies['Release_Date'] = movies['Release_Date'].astype(int)
movies['Release_Date'] = movies['Release_Date'].astype(str)
movies.head()
```

```python
movies = pd.DataFrame(movies[movies['Release_Date'].str.len() > 4])
```

```python
movies['Year'] = pd.DataFrame(movies['Release_Date'].str[:4])
```

```python
movies['Month'] = pd.DataFrame(movies['Release_Date'].str[4:6])
movies.head()
```

```python
movies2012_2015 = movies[(movies['Year']>='2012') & (movies['Year']<='2015')] # filtering 2012-2017 movies
```

```python
movies2012_2015['Count'] = 1
movies2012_2015_Grouped = movies2012_2015.groupby(['Year','Month'])['Count'].sum().reset_index()
movies2012_2015_Grouped.head()
```

```python
year2012 = movies2012_2015_Grouped[movies2012_2015_Grouped['Year']=='2012']
year2013 = movies2012_2015_Grouped[movies2012_2015_Grouped['Year']=='2013']
year2014 = movies2012_2015_Grouped[movies2012_2015_Grouped['Year']=='2014']
year2015 = movies2012_2015_Grouped[movies2012_2015_Grouped['Year']=='2015']
```

```python
# plotting the number of movies released
x_2012 = year2012['Month'].astype(float)
x_2013 = year2013['Month'].astype(float)
x_2014 = year2014['Month'].astype(float)
x_2015 = year2015['Month'].astype(float)
y_2012 = year2012['Count']
y_2013 = year2013['Count']
y_2014 = year2014['Count']
y_2015 = year2015['Count']
months = ['January','Febuary','March','April', 'May', 'June', 'July', 'August', 'September', 'October','November', 'December']
x = np.arange(len(months))
bar_width = 0.2
opacity = 0.4 #not so dark
plt.bar(x, y_2012,bar_width, color='#ac5353', label='2012', alpha=opacity)
plt.bar(x+bar_width, y_2013,bar_width, color='#ff6666', label='2013', alpha=opacity)
plt.bar(x+bar_width+bar_width, y_2014,bar_width, color='#40ff00', label='2014', alpha=opacity)
plt.bar(x+bar_width+bar_width+bar_width, y_2015,bar_width, color='#ff00bf', label='2015', alpha=opacity)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.05),
          ncol=2, fancybox=True, shadow=True)
plt.xlabel('Months')
plt.ylabel('Number of Movies Released')
plt.title("Number of movies released in USA [2012-2015]")
plt.xticks(x,months, rotation=45)
plt.xlim(0,12)
plt.ylim(0,60)
plt.tight_layout()
plt.show()
```

# Good Movies Vs Bad Movies

```python
movies["Count"]=1
```

```python
rating_grouped = pd.DataFrame(movies.groupby(['Rating'], sort = True)['Count'].sum().reset_index()) # grouping by month for manhattand data and
```

```python
rating_grouped = rating_grouped[rating_grouped['Rating']>0.0]
rating_grouped.head()
```

```python
# This is  the colormap.
cm = plt.cm.get_cmap('RdYlBu_r')
n, bins, patches = hist(rating_grouped['Rating'], weights=rating_grouped['Count'],normed=1, color='green')
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# scale values to interval [0,1]
col = bin_centers - min(bin_centers)
col /= max(col)

for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))
plt.title('Good Movies Vs Bad Movies');
plt.xlabel('Rating of Movies');
plt.ylabel('Frequency of Ratings');
#plt.ylim(0,100)
plt.show()
#plt.ylim(0,50); #% Scale x axis manually.
```

# Top 3 movies from the released movies [2012-2015]

```python
ratingYear = movies[['Rating', 'Year', 'UrlPoster', 'Title']]
ratingSorted = ratingYear.sort('Rating', ascending=False)
topMovies = ratingSorted.head(5)
topMovies
```

```python
Image= "Image"
Image_path = os.path.join(path,Image)
for each in range(0,3):
    url = topMovies.UrlPoster.iloc[each]
    title = topMovies.Title.iloc[each]
    Image_title= os.path.join(Image_path,title)
    time=datetime.datetime.now()
    time=time.strftime("%y-%m-%d-%H-%M")
    name = Image_title+time
    with urllib.request.urlopen(url) as url:
        with open(name+'.jpg', "wb") as f:
            f.write(url.read())           
```
