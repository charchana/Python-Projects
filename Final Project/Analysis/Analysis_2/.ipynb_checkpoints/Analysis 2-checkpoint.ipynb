# Analysing IMDB Rating Distribution for each country

```python
import os
import pandas as pd
import calendar
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import hist
import datetime
import urllib.request
```

```python
path = os.getcwd() # getting the path of current working directory to create the folder structure
#configfile = os.path.join(path + '\Data\IMDB_Movies.csv') # getting the path of the json files to be read
```

```python
configfile = os.path.join(path + '\Data\Movies_Gross.csv') # getting the path of the json files to be read
```

```python
#fields = ['DATE', 'BOROUGH']
all_movies = pd.DataFrame.from_csv(configfile, encoding = "ISO-8859-1").reset_index()
all_movies
all_movies = all_movies.replace('None','0')
all_movies.head()
```

```python
all_moviesRatingcountry = pd.DataFrame(all_movies,columns = ['Country', 'Rating'])
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry.notnull()]
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='0']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='except']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='Worldwide']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='worldwide']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='Pty']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='opening']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='New']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Country']!='Films']
all_moviesRatingcountry = all_moviesRatingcountry[all_moviesRatingcountry['Rating']<=10]
all_moviesRatingcountry.head()
```

```python
g = sns.factorplot(kind='box',        # Boxplot
               y='Rating',       # Y-axis - values for boxplot
               x='Country',        # X-axis - first factor
               data=all_moviesRatingcountry,        # Dataframe 
               size=10,            # Figure size (x100px)      
               aspect=1.55,
               color="#00ffbf",# Width = size * aspect 
               legend_out=False)  # Make legend inside the plot
g.set_xticklabels(rotation=90)
plt.title('Rating Distribution for each Country 2012-2015', fontsize=18)
plt.ylim(2, 9)
plt.show()
```

# Profits  of USA  UK and Worldwide

```python
all_movies['Gross']= pd.to_numeric(all_movies['Gross'])
```

```python
Gross_Grouped = pd.DataFrame(all_movies.groupby(['IMDB_ID','Title','Country','Rating'], sort = True)['Gross'].sum().reset_index()) # grouping by month for manhattand data and
```

```python
Gross_Grouped = Gross_Grouped[Gross_Grouped['Country']!='0']
Gross_Grouped = Gross_Grouped[Gross_Grouped['Country']!='except']
```

```python
Gross_Grouped_USA = Gross_Grouped[Gross_Grouped['Country']=='USA']
Gross_Grouped_UK = Gross_Grouped[Gross_Grouped['Country']=='UK']
Gross_Grouped_Wordwide = Gross_Grouped[(Gross_Grouped['Country']!='USA') & (Gross_Grouped['Country']!='UK')]
#Gross_Grouped_UK = Gross_Grouped_UK[['Gross', 'Rating']]
#Gross_Grouped_Wordwide = Gross_Grouped_Wordwide[['Gross', 'Rating']]

# sorting the top 10 movies of highest profit in USA
Gross_Grouped_USA = Gross_Grouped_USA[['Title','Gross', 'Rating']]
sortedUSA = Gross_Grouped_USA.sort('Gross', ascending=False)
sortedUSA = sortedUSA.head(10)
sortedUSA['Rating'] = sortedUSA['Rating'].astype(str)
sortedUSA['Gross'] = sortedUSA['Gross'].astype(str)

# sorting the top 10 movies of highest profit in UK
Gross_Grouped_UK = Gross_Grouped_UK[['Title','Gross', 'Rating']]
sortedUK = Gross_Grouped_UK.sort('Gross', ascending=False)
sortedUK = sortedUK.head(10)
sortedUK['Rating'] = sortedUK['Rating'].astype(str)
sortedUK['Gross'] = sortedUK['Gross'].astype(str)

# sorting the top 10 movies of highest profit in worldwide
Gross_Grouped_WorldWide = Gross_Grouped_Wordwide[['Title','Gross', 'Rating']]
sortedWorldwide = Gross_Grouped_WorldWide.sort('Gross', ascending=False)
sortedWorldwide = sortedWorldwide.head(10)
sortedWorldwide['Rating'] = sortedWorldwide['Rating'].astype(str)
sortedWorldwide['Gross'] = sortedWorldwide['Gross'].astype(str)
```

```python
#Gross_Grouped_Wordwidess = sortedUSA[sortedUSA['Title'] + str(sortedUSA['Rating'])].reset.index()
sortedUSA['Period'] = sortedUSA[['Title', 'Rating', 'Gross']].apply(lambda x: ','.join(x), axis=1)
sortedUK['Period'] = sortedUK[['Title', 'Rating', 'Gross']].apply(lambda x: ','.join(x), axis=1)
sortedWorldwide['Period'] = sortedWorldwide[['Title', 'Rating', 'Gross']].apply(lambda x: ','.join(x), axis=1)
sortedUK
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mpld3

fig = plt.figure(figsize=(14,14))
N=10
#first plot
fig.add_subplot(2,2,1)
scatter1 = plt.scatter(sortedUSA[['Rating']],
                     sortedUSA['Gross'],
                     s=1000 * np.random.random(size=N),
                     alpha=0.3, color = 'green',
                     cmap=plt.cm.jet)
plt.grid(color='white', linestyle='solid')
plt.title("Top 10 Movies of USA with highest profit ", size=18)
plt.xlim(1,10)
plt.xlabel('Rating', rotation=45, size=14)
plt.ylabel('Gross', rotation=90, size=14)
labels1 = [i for i in sortedUSA['Period']]
tooltip1 = mpld3.plugins.PointLabelTooltip(scatter1, labels=labels1)
mpld3.plugins.connect(fig, tooltip1)


# second plot
fig.add_subplot(2,2,2)
scatter2 = plt.scatter(sortedUK[['Rating']],
                     sortedUK['Gross'],
                     s=1000 * np.random.random(size=N),
                     alpha=0.3, color = 'blue',
                     cmap=plt.cm.jet)
plt.title("Top 10 Movies of UK with highest profit ", size=18)
plt.xlim(1,10)
plt.xlabel('Rating', rotation=45, size=14)
plt.ylabel('Gross', rotation=90, size=14)
labels2 = [i for i in sortedUK['Period']]
tooltip2 = mpld3.plugins.PointLabelTooltip(scatter2, labels=labels2)
mpld3.plugins.connect(fig, tooltip2)


#third plot
fig.add_subplot(2,2,3)
#plt.axis('equal')
scatter3 = plt.scatter(sortedWorldwide[['Rating']],
                     sortedWorldwide['Gross'],
                     s=1000 * np.random.random(size=N),
                     alpha=0.3, color = 'magenta',
                     cmap=plt.cm.jet)
plt.title("Top 10 Movies of Worldwide with highest profit ", size=18)
plt.xlim(1,10)
plt.xlabel('Rating', rotation=45, size=14)
plt.ylabel('Gross', rotation=90, size=14)
labels3 = [i for i in sortedWorldwide['Period']]
tooltip3 = mpld3.plugins.PointLabelTooltip(scatter3, labels=labels3)
mpld3.plugins.connect(fig, tooltip3)
mpld3.display()
```

## Movie Titanic Reached Highest Popularity Worldwide

```python
imdb_movie = os.path.join(path + '\Data\IMDB_Movies.csv') # getting the path of the json files to be read
```

```python
titanic = pd.DataFrame.from_csv(imdb_movie, encoding = "ISO-8859-1").reset_index()
titanic = pd.DataFrame(titanic,columns = ['Title', 'UrlPoster'])
titanic = pd.DataFrame(titanic[titanic['Title'] == 'Titanic'])
titanic
```

```python
Image= "Image"
Image_path = os.path.join(path,Image)
url = titanic.UrlPoster.iloc[0]
title = titanic.Title.iloc[0]
Image_title= os.path.join(Image_path,title)
time=datetime.datetime.now()
time=time.strftime("%y-%m-%d-%H-%M")
name = Image_title+time
with urllib.request.urlopen(url) as url:
    with open(name+'.jpg', "wb") as f:
        f.write(url.read()) 
```
