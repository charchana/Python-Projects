```python
#fields = ['DATE', 'BOROUGH']
all_movies = pd.DataFrame.from_csv(configfile, encoding = "ISO-8859-1").reset_index()
all_movies
all_movies = all_movies.replace('None','0')
all_movies
```

```python
all_movies['Gross']= pd.to_numeric(all_movies['Gross'])
```

```python
Gross_Grouped = pd.DataFrame(all_movies.groupby(['IMDB_ID','Title','Country','Rating'], sort = True)['Gross'].sum().reset_index()) # grouping by month for manhattand data and
```

```python
import numpy as np                          
import matplotlib.pyplot as plt            
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.animation as animation

# Number of particles
numP = 2
# Dimensions
DIM = 3
timesteps = 2000

# with open('//home//data.dat', 'r') as fp:
#     particleData = []
#     for line in fp:
#         line = line.split()
#         particleData.append(line)

columns = ['Country', 'Rating', 'Gross']
df = pd.DataFrame(Gross_Grouped, columns = columns)

x = df['Country']
#xArray = np.array(x, dtype=pd.Series)
y = df['Rating']
z = df['Gross']     

# Attaching 3D axis to the figure
fig = plt.figure()
ax = p3.Axes3D(fig)

# Setting the axes properties
border = 1
ax.set_xlim3d([-border, border])
ax.set_ylim3d([-border, border])
ax.set_zlim3d([-border, border])


def animate(i):
    print("entering animate")
    global x, y, z, numP
    #ax.clear()
    ax.set_xlim3d([-border, border])
    ax.set_ylim3d([-border, border])
    ax.set_zlim3d([-border, border])
    idx0 = i*numP
    idx1 = numP*(i+1)
    ax.scatter(x,y,z)

ani = animation.FuncAnimation(fig, animate, frames=timesteps, interval=1, blit=False, repeat=False)
plt.show()
```
