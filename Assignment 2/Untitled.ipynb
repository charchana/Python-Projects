```python
import nltk 
import string
```

```python
from nltk.corpus import gutenberg
from nltk.probability import FreqDist
from operator import itemgetter
```

```python
words = nltk.corpus.gutenberg.words()
len(words)
```

```python
wordsLower =[]
for e in words:
    wordsLower.append(e.lower())
```

```python
exactWords = [''.join(pun for pun in s if pun not in string.punctuation) for s in wordsLower] # removing the punctuations from the list
exactWords = [s for s in exactWords if s]
```

```python
totalNoWords = len(exactWords)
```

```python
freq = FreqDist(exactWords)
```

```python
sortedFreq = sorted(freq.items(), key=itemgetter(1),reverse=True) 
sortedFreq
```

```python
listOfWords = [list(elem) for elem in sortedFreq]
listOfWords
```

```python
for count,elem in enumerate(listOfWords,1):
    #print(count,elem)
       elem.append(count)
        
```
